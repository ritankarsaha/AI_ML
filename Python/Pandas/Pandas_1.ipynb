```python
import pandas as pd
import tabula
pdf_path = 'UG2023 final alloted courses.pdf'

# Read the PDF file into a list of DataFrames
df_list = tabula.read_pdf(pdf_path, pages='all')

# Check the length of df_list to understand how many tables were found
print(f"Number of tables extracted: {len(df_list)}")

# Display each DataFrame individually to check their contents
for i, df in enumerate(df_list):
    print(f"Table {i+1}:")
    display(df)

# Concatenate all DataFrames into a single DataFrame (if multiple tables are present)
if len(df_list) > 1:
    combined_df = pd.concat(df_list, ignore_index=True)
else:
    combined_df = df_list[0]

# Display the concatenated DataFrame
combined_df
```

```python
# Clean the DataFrame by removing line breaks in the 'Details of Register Courses for Sem-III' and 'Name' columns
combined_df['Details of Register Courses for Sem-III'] = combined_df['Details of Register Courses for Sem-III'].str.replace('\r', ' ', regex=False)
combined_df['Name'] = combined_df['Name'].str.replace('\r', ' ', regex=False)

# Ensure the columns exist and filter the DataFrame
if 'Details of Register Courses for Sem-III' in combined_df.columns and 'Name' in combined_df.columns and 'Roll No.' in combined_df.columns:
    # Filter the DataFrame for the specific course name
    filtered_df = combined_df[combined_df['Details of Register Courses for Sem-III'].str.contains('OE2E01 Introduction to Sensors and Actuators', na=False)]
    
    # Select the relevant columns
    selected_columns_df = filtered_df[['Details of Register Courses for Sem-III', 'Name', 'Roll No.']]
    
    # Set display options to show all rows
    pd.set_option('display.max_rows', selected_columns_df.shape[0])
    
    # Display the filtered DataFrame
    display(selected_columns_df)
else:
    print("One or more columns not found in the DataFrame")
```
